```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
mu, sigma, n = 5, 0.3, 1000
first = np.random.normal(mu, sigma, n)

mu, sigma, n = 10, 0.2, 1000
second = np.random.normal(mu, sigma, n)
```

```python
def plot_norm(counts, bins):
    plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi))*np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=3, color='r')
    return None
```

```python
plt.figure(figsize=(12,6))

plt.subplot(121)
plt.ylim((0,3))
count, bins, ignore = plt.hist(first, 50, density=True)
plot_norm(count, bins)

plt.subplot(122)
plt.ylim((0,3))
count, bins, ignore = plt.hist(second, 50, density=True)
plot_norm(count, bins)


plt.show()
```
